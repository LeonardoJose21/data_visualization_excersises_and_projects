
# Data Viz with Seaborn

## The Data

I'll be working with the <b>titanic</b> dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(data=titanic, x='fare',y='age')
```

```python
sns.histplot(data=titanic, x='fare', color='red', bins=30)
```

```python
sns.boxplot(data = titanic, x='class', y='age')
```

```python
sns.swarmplot(data = titanic, x='class', y='age', palette= 'rainbow')
```

```python
sns.countplot(titanic, x='sex')
```

** Correlation between each variable**

```python
sns.heatmap(titanic.corr(), cmap='coolwarm',annot=True)
plt.title('Correlation ')
```

```python
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
```
