
# Pandas Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
plt.style.use('bmh')
```

** Creating a scatter plot of 'b' vs 'a' **

```python
df3.plot(kind='scatter',x='a',y='b', c='r', figsize=(10,3))
```

** Creating a histogram of the 'a' column.**

```python
df3['a'].plot.hist(bins=30,color='r')
```

** Creating a boxplot comparing the a and b columns.**

```python
df3[['a','b']].plot.box()
```

** Creating a kde plot of the 'd' column **

```python
df3['d'].plot.kde(linewidth=2, linestyle='dashed')
```

** Creating an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
df3.plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
