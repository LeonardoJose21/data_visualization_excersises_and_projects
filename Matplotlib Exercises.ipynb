# Matplotlib Practice


### Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt **

```python
import matplotlib.pyplot as plt
```


* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

# Add set of axes to figure
ax = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)
# Plot on that set of axes
ax.plot(x, y, 'b')
ax.set_xlabel('x') # Notice the use of set_ to begin methods
ax.set_ylabel('y')
ax.set_title('Plot title')
```


** <ul> <li>Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.</li>
    <li>Plot (x,y) on both axes. And call the figure object to show it</li>.</ul>**

```python
fig_2 = plt.figure()
# Add set of axes to figure
ax_1 = fig_2.add_axes([0, 0, 1, 1]) 
ax_2 = fig_2.add_axes([0.2, 0.5, .2, .2]) 
ax_1.plot(x,y,'r--')
ax_2.plot(x,y,'r-.')
```

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4] Then use x,y, and z arrays. there are x limits and y limits on the inserted plot**

```python
fig_3 = plt.figure()
# Add set of axes to figure
ax_1_3 = fig_3.add_axes([0, 0, 1, 1]) 
ax_2_3 = fig_3.add_axes([0.2, 0.5, .4, .4]) 
ax_1_3.plot(x,z,'b:')
ax_2_3.plot(x,y,'r-.')

ax_2_3.set_ylim([30, 50])
ax_2_3.set_xlim([20, 22])

ax_1_3.set_xlabel('x') 
ax_1_3.set_ylabel('z')

ax_2_3.set_xlabel('x') 
ax_2_3.set_ylabel('y')
ax_2_3.set_title('zoom')
```



** Use plt.subplots**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))
axes[0].plot(x,y, 'b:')
axes[1].plot(x,z,'r*-', linewidth=0.1)
plt.tight_layout()
```
